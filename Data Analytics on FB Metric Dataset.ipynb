### #1  Creating Sub-Datasets
#### Remember:  
* In Python the Upper Bound is Always neglected
* Attributes that are Compulsory, should never be left blank
* Appropriate Range must be given wherever applicable

```python
ds1 = dataset.iloc[:, 0:4]  # All Rows and Columns from 0 to 3
print (ds1)
ds2 = dataset.iloc[0:4, :]  # Rows from 0 to 3 and All Columns
print (ds2)
ds3 = dataset.iloc[1:5, 1:7]  # Rows from 1 to 4 and 1 to 6 Columns
print (ds3)
typ = dataset.iloc[:, 1]  # All Rows and Column no. 1
print (typ)
```

#### The above statements would've been slightly difficult to identify the outputs. The better way is you try it out one at a time.

### #2 Merge Data

The first line of code below is slightly off-beat to understand, (take a look at it, and return) so let me brief it. In there I'm asking S1 to collect all data from dataset where Status is set to 'Status' (and not 'Posts'). For same I'm asking the help of a Column that I've already created above, that is `typ`

#### What are we learning new?
* pd.concat - Using which we will be concatenating 2 dataframes in different ways
* pd.Series - One of the Datatype besides Dataframe available in Pandas (We are using this just to create Index numbers)

```python
S1 = dataset[typ == 'Status']
print("S1:\n", S1)

ds4 = dataset.iloc[:, [3, 4, 5, 7]]
print("\n DS4:\n", ds4)

StatusDetails = pd.concat([S1, ds4])  # This DF will contain all details of Status and NaN values from Posts
print("This data is not well arranged and has NaN values in it\n", StatusDetails.head(50))

indexSeries = pd.Series(range(1, dataset.__len__()+1))  # Creating a Series of Index numbers
print("These are just Index numbers:\n", indexSeries.head(50))

dataIdx = pd.concat([indexSeries, dataset], axis=1)  # Merging data of Indexes to Dataset
print("The dataset now has Index numbers combined that we just created!\n", dataIdx)
```

### #3 Sort Data
Time to sort the available data. Sorting is again very simple. You don't have to worry about any sorting methods, you just need to call the sorting upon parameters you wish. For this we'll create a simple dataset where dataset is simple to understand in context of Dataset.

```python
unsortedDF = pd.DataFrame(np.random.randn(10, 2), index=[1, 4, 6, 2, 3, 5, 9, 8, 0, 7], columns=['col2', 'col1'])
print("Dataset created:", unsortedDF)

sortedAscDF = unsortedDF.sort_index()  # Ascending Arrangement by default
print("Simple Sorting: Default Ascending\n", sortedAscDF)

sortedDecDF = unsortedDF.sort_index(ascending=False)  # Descending by default
print("Simple Sorting: Descending\n", sortedDecDF)

sortedColDF = unsortedDF.sort_index(axis=1)  # Column Sortprint | Axis 1 is Column and 0 is row
print("Simple Sorting: Column wise\n", sortedColDF)
```

ðŸ˜ƒ Well Done!
Now that you understand that Sorting and Data is not a big deal to handle, here's a bonus sorting method. If you are still struggling with above, you can improve with bit of practice. If you are ready for bonus and scroll and learn it Junior Data Scientist!

#### Bonus Method of Sorting!
This will enable you to sort on multiple values. The trick is simple that you need to pass on List. Let's see how:

```python
# Selective Sort
sortedByValue = unsortedDF.sort_values(by=['col1', 'col2'])
print(sortedByValue)
```

Dataset that we will be using is:

`In    0     Type  Category
0    1     Photo         2
1    2    Status         2
2    3     Photo         3
3    4     Photo         2
4    5     Photo         2
5    6    Status         2

```python
newDataSet = dataIdx.iloc[0:10, [0, 2, 3]]
print(newDataSet,"\n")

meltDS = pd.melt(newDataSet, id_vars=[0], var_name='Attribute', value_name='Values')
print("\n MELTED TO LONG FORMAT \n", meltDS)

wideDS = meltDS.pivot(index=0, columns='Attribute', values='Values')
print("\n CASTING TO WIDE FORMAT \n", wideDS)
```
